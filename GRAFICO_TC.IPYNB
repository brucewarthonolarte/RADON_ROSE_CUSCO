```python
#CODIGO FINAL PARA TC_PRIMERAS 4 SEMANAS (GRAFICO DE DISPERSIÃN CON LINEAS RECTAS Y MARCADORES)
import pandas as pd
import matplotlib.pyplot as plt

# Ruta del archivo Excel
archivo_excel = r"C:\Users\Bruce\Desktop\ARTICULOS\ARTICULO PUCP\RADON\LIMPIEZA_CUSCO\TC\primeros_4_periodos_tc_solo_radon_16_11.xlsx"

# Cargar el archivo Excel en un DataFrame de Pandas
df = pd.read_excel(archivo_excel)

# Imprimir informaciÃ³n para debug
print(f"NÃºmero de filas: {len(df)}")
print(f"Contenido del DataFrame:\n{df.head()}")

# Obtener el nÃºmero de filas y columnas en el DataFrame
num_filas, num_columnas = df.shape

# Configurar subplots
fig, axs = plt.subplots(4, 3, figsize=(15, 15))  # 4 filas, 3 columnas

# Iterar sobre las filas y columnas
for i in range(4):
    for j in range(3):
        # Calcular el Ã­ndice correspondiente en el DataFrame
        index = i * 3 + j

        # Verificar si hay mÃ¡s Ã­ndices que filas en el DataFrame
        if index >= num_filas:
            axs[i, j].axis('off')  # Desactivar el eje si no hay datos para ese subplot
        else:
            # Obtener datos para el eje X
            x_data = list(range(1, 5))  # Utilizar valores enteros

            # Obtener datos para el eje Y y convertir a tipo numÃ©rico
            y_data = pd.to_numeric(df.iloc[index, 2:], errors='coerce')  # Tomar todas las columnas a partir de la tercera

            # Obtener el nombre del pozo para usarlo como tÃ­tulo
            nombre_pozo = df.iloc[index, 0]

            # Graficar dispersiÃ³n con lÃ­neas rectas y marcadores
            axs[i, j].plot(x_data, y_data, marker='o', linestyle='-')

            # Configurar etiquetas y tÃ­tulo
            axs[i, j].set_xticks(x_data)  # Establecer los ticks en los valores enteros
            axs[i, j].set_xlabel('Periodo')
            axs[i, j].set_ylabel('ConcentraciÃ³n de RadÃ³n (Bq/m3)')
            axs[i, j].set_title(f'Pozo {nombre_pozo}')
# Agregar tÃ­tulo general

plt.suptitle("Twin chamber - Primeros cuatro periodos - Cusco", fontsize=16)

# Ajustar diseÃ±o y mostrar grÃ¡ficos
plt.tight_layout()
plt.show()
```
